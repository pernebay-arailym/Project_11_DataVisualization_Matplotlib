Load Necessary Libraries

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)

plt.title('Our First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'Arial'}) 
plt.ylabel('Y Axis')

plt.show()
```
